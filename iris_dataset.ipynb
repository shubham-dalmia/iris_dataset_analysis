```python
sum_data = df["sepal length (cm)"].sum()
mean_data = df["sepal length (cm)"].mean()
median_data = df["sepal length (cm)"].median()
 
print("Sum:",sum_data, "\nMean:", mean_data, "\nMedian:",median_data)
```

```python
min=df['sepal length (cm)'].min()
max=df['sepal length (cm)'].max()
print("Minimum sepal length =",min, "\nMaximum =",max)
```

```python
fig, ax = plt.subplots()
x_index = 0
colors = ['blue', 'red', 'green']

for label, color in zip(range(len(iris.target_names)), colors):
    ax.hist(iris.data[iris.target==label, x_index], 
            label=iris.target_names[label],
            color=color)

ax.set_xlabel(iris.feature_names[x_index])
ax.legend(loc='upper right')
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

x_index = 0
y_index = 1

colors = ['blue', 'red', 'green']

for label, color in zip(range(len(iris.target_names)), colors):
    ax.scatter(iris.data[iris.target==label, x_index], 
                iris.data[iris.target==label, y_index],
                label=iris.target_names[label],
                c=color)

ax.set_xlabel(iris.feature_names[x_index])
ax.set_ylabel(iris.feature_names[y_index])
ax.legend(loc='upper left')
plt.show()
```
